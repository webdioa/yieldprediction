```python
import pandas as pd
import numpy as np
```

```python
#df = pd.read_excel(r"C:\web\data\keras\corn_data\copy alldata.xlsx" ,index_col="ID_Number")
original_df = pd.read_excel(r"C:\web\data\corn\All Data with description.xlsx" ,sheet_name="All data",usecols="A,Z,AG,AJ")
original_df.tail(5)
```

```python
df_nona = original_df.dropna()
df = df_nona.drop(labels=0, axis=0)
df = df.rename(columns={'Unnamed: 0': 'ID_Number', 'Unnamed: 25': 'planting_date', 'Unnamed: 32': 'harvest_date', 'Unnamed: 35': 'yield/hec'})
df = df.set_index('ID_Number')
df['planting_date']= pd.to_datetime(df['planting_date'])
df['harvest_date']= pd.to_datetime(df['harvest_date'])
#df = df.astype({'planting_date': int, 'harvest_date': complex, 'yield/hec':})
df['total_planting_days'] = (df['harvest_date'] - df['planting_date']) / np.timedelta64(1, 'D')
df["image_date1"] =  pd.Timestamp('2019-01-06')
df["image_date2"] =  pd.Timestamp('2019-02-28')
df['total_days_image_1'] = (df['image_date1'] - df['planting_date']) / np.timedelta64(1, 'D')
df['total_days_image_2'] = (df['image_date2'] - df['planting_date']) / np.timedelta64(1, 'D')

df['dt_im2_har'] = (df['harvest_date'] - df['image_date2']) / np.timedelta64(1, 'D')
df
```

```python
df[df["dt_im2_har"]<0].index # ID ที่ วันเก็บเกี่ยวมาก่อนถ่ายภาพที่ 2 ภาพควรจะดูไม่มี crop 
# อันที่ยังมี crop อยู่ 7 11 13 28 29 50 (มีส่วนหนึ่งเป็นเขียว) 86 151
```
